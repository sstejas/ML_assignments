Q1. Consider the given list as [2, 7, 4, 1, 3, 6]. Write a program to count pairs of elements with
sum equal to 10.

```python
def count_pairs(lst,target = 10):
  count = 0
  for i in range(len(lst)):
    for j in range(i,len(lst)):
      if lst[i]+lst[j] == target:
        count+=1
  return count

lst = [2,7,4,1,3]
print("The total count is ",count_pairs(lst))
```

Q2.Write a program that takes a list of real numbers as input and returns the range (difference
between minimum and maximum) of the list. Check for list being less than 3 elements in
which case return an error message (Ex: “Range determination not possible”). Given a list
[5,3,8,1,0,4], the range is 8 (8-0).



```python
def diff(lst):
  if(len(lst)<3):
    raise Exception("Error : List should consist more than 3 elements ")
  else:
    diff = max(lst) - min(lst)
    return diff

n = int(input("Enter the no.of elements : "))
lst = []
for i in range(n):
  lst.append(int(input(f"Enter the element {i} : ")))

print("The list you entered : ",lst)
print(diff(lst))
```

Q3.Write a program that accepts a square matrix A and a positive integer m as arguments and
returns A**m

```python
def positive(m):
  if(m<0):
    raise Exception("Error : the number is negative ")
  else:
    return m

def power_matrix(A,m):
  if(len(A)!=len(A[0]) ):
    raise Exception("Error : the matrix should consis same no.of rows and colums")
  else:
    power_matrix=[[0 for i in range(len(A))] for _ in range(len(A))]
    for i in range(len(A)):
      for j in range(len(A)):
        power_matrix[i][j]=A[i][j]**m
    return power_matrix
A=[[2,3],[3,2]]
try:
  try:
    m=int(input("Enter the number : "))
    m=positive(m)
  except Exception as e:
    print(e)

  print("the power matrix is ",power_matrix(A,m))
except Exception as e:
  print(e)
```

4.Write a program to count the highest occurring character & its occurrence count in an input
string. Consider only alphabets. Ex:  for “hippopotamus” as input string, the maximally
occurring character is ‘p’ & occurrence count is 3.

```python
def max_occurence(word):
  occurence = {}
  for i in word:
    if i in occurence:
      occurence[i]+=1
    else:
      occurence.update({i:1})
  max = 0
  ele = ""
  for i in occurence:
    if occurence[i] > max:
      max = occurence[i]
      ele = i

  return ele,max

word = input("Enter the word : ")
print(max_occurence(word))
```
